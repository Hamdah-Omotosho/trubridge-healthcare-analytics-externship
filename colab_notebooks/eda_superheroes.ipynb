```python
import pandas as pd

# Load the CSV
heroes_df = pd.read_csv("heroes_information.csv")

# Preview the first few rows
heroes_df.head()
```

```python
# Basic info
heroes_df.info()

# Summary statistics
heroes_df.describe()

# Check for missing values
print("Missing values:\n", heroes_df.isnull().sum())

# Histogram of Height
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(heroes_df['Height'], bins=20, kde=True)
plt.title('Distribution of Superhero Heights')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()

# Boxplot to detect outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=heroes_df['Height'])
plt.title('Boxplot of Superhero Heights')
plt.show()
```

```python
# Central tendency of Height
print("Mean Height:", heroes_df["Height"].mean())
print("Median Height:", heroes_df["Height"].median())
print("Mode Height:", heroes_df["Height"].mode()[0])

# Check most common alignment
heroes_df["Alignment"].value_counts()
```

```python
# Fill missing Height with median (if needed)
heroes_df["Height"].fillna(heroes_df["Height"].median(), inplace=True)

# Optional: drop rows with too many missing values
heroes_df.dropna(subset=["Weight"], inplace=True)
```

```python
# Boxplot for Weight
plt.figure(figsize=(8,5))
sns.boxplot(x=heroes_df["Weight"])
plt.title("Boxplot of Superhero Weights")
plt.show()
```

```python
# Scatter plot: Height vs. Weight
plt.figure(figsize=(8,5))
sns.scatterplot(x=heroes_df['Height'], y=heroes_df['Weight'])
plt.title("Height vs. Weight of Superheroes")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()

# Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(heroes_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot: Weight by Publisher
plt.figure(figsize=(8,5))
sns.boxplot(x=heroes_df["Publisher"], y=heroes_df["Weight"])
plt.title("Weight Distribution: Marvel vs DC")
plt.xlabel("Publisher")
plt.ylabel("Weight")
plt.show()

```
