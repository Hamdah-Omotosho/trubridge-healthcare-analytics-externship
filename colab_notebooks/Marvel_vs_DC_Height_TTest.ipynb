```python
# Step 1: Import libraries
import pandas as pd
from scipy import stats
from google.colab import files

# Step 2: Upload your CSV file
uploaded = files.upload()

# Step 3: Replace with the correct filename (exact match)
df = pd.read_csv('heroes_information.csv')  #

# Step 4: Preview the data
print(df.head())
print(df.columns)
```

```python
# Drop rows with missing Height or Publisher
df = df[['Publisher', 'Height']].dropna()

# Filter for Marvel and DC
df = df[df['Publisher'].isin(['Marvel Comics', 'DC Comics'])]

# Separate the two groups
marvel_heights = df[df['Publisher'] == 'Marvel Comics']['Height']
dc_heights = df[df['Publisher'] == 'DC Comics']['Height']

# Run T-test
t_stat, p_value = stats.ttest_ind(marvel_heights, dc_heights)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("✅ The difference in height is statistically significant.")
else:
    print("❌ No significant difference in height between Marvel and DC superheroes.")
```
