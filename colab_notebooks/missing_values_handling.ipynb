```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno

# Sample dataset with missing values
data = {
    'Patient_ID': [101, 102, 103, 104, 105, 106],
    'Age': [25, np.nan, 40, 35, np.nan, 55],
    'Blood Pressure': [120, 130, np.nan, 110, 125, 135],
    'Hospital Stay (Days)': [np.nan, 5, 3, np.nan, 7, 4]
}
df = pd.DataFrame(data)
df
```

```python
print("Missing Value Matrix:\n", df.isnull())
print("\nTotal Missing per Column:\n", df.isnull().sum())
```

```python
msno.matrix(df)
plt.show()
```

```python
missing_percent = df.isnull().sum() / len(df) * 100
print("Percentage of Missing Values:\n", missing_percent)
```

```python
df_dropped = df.dropna()
```

```python
df_filled = df.copy()
df_filled['Age'].fillna(df['Age'].mean(), inplace=True)
df_filled['Blood Pressure'].fillna(df['Blood Pressure'].median(), inplace=True)
df_filled['Hospital Stay (Days)'].fillna(df['Hospital Stay (Days)'].mean(), inplace=True)
```

```python
df_filled.to_csv('handled_missing_values.csv', index=False)
```
